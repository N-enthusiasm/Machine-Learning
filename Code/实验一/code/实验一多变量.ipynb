```python
#
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

#
train_data=pd.read_csv('ex1data2.txt',header=None,names=['Size','Bedrooms','Price'])

#
train_data=(train_data-train_data.mean())/train_data.std()


#
def compute_loss(X,y,theat):
    m=float(2*len(X))
    count=np.power((X*theat.T-y),2)
    return np.sum(count)/m

#
def gadient_fun(X,y,theat,alpha,iteators):
    temp=np.matrix(np.zeros(theat.shape))#
    parameters = int(theat.ravel().shape[1])#ravel ,
    cost = np.zeros(iters)#iters0,
    
    for i in range(iters):
        error = (X * theat.T)-y
        for j in range(parameters):
            term = np.multiply(error,X[:,j])#hx-yx
            temp[0,j] = theat[0,j] - ((alpha/len(X)) * np.sum(term))
            
        theat = temp
        cost[i] = compute_loss(X,y,theat)
        print('Iters = ',i,'loss =',cost[i])
        
    return theat,cost

alpha = 0.01 #
iters =1000 #

#'Ones'1
train_data.insert(0,'0nes',1)

#XY
cols = train_data.shape[1] #
X = train_data.iloc[:,0:cols-1]
y = train_data.iloc[:,cols-1:cols]

X=np.matrix(X.values)
y=np.matrix(y.values)
theat=np.matrix(np.array([0,0,0]))

#
theat,cost=gadient_fun(X,y,theat,alpha,iters)

#
x1=np.linspace(X[:,1].min(),X[:,1].max(),100)
x2=np.linspace(X[:,2].min(),X[:,2].max(),100)
x1,x2=np.meshgrid(x1,x2)
f = theat[0,0]+theat[0,1]*x1+theat[0,2]*x2

fig = plt.figure()
Ax = Axes3D(fig)
Ax.plot_surface(x1, x2, f, rstride=1, cstride=1, cmap=cm.viridis,label='prediction')

Ax.scatter(X[:,1],X[:,2],y[:100,0],c='r')

plt.show()

#
plt.plot(np.arange(iters),cost,'r')
plt.show()

    
X_new=np.matrix([4500,3])
#X_new=(X_new-X_new.mean())/X_new.std()
result=theat[0,0]+theat[0,1]*X_new[:,0]+theat[0,2]*X_new[:,1]

print(result)
```
